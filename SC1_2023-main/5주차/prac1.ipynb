신용카드 신청자들에 대해서 해당 신청의 승인 여부를 기록한 두 데이터를 merge한 후 전처리하고자 합니다.
- 신청 관련 데이터: application_record.csv.zip / 신용 관련 데이터: credit_record.csv.zip
- 출처: Kaggle, https://www.Kaggle.com/datasets/rikdifos/credit-card-approval-prediction

 - ID 컬럼 기준으로 merge
 - 결측치 Imputing: 범주형 변수의 nan 값은 others로 변경
 - STATUS 컬럼의 C와 X를 각각 6과 7로 변경
 - 수치형 변수들에 대해 Scaling
 - 전체 데이터 Train, Test로 8:2로 파티셔닝

```python
import pandas as pd 

ar = pd.read_csv("data/application_record.csv")
cr = pd.read_csv("data/credit_record.csv")
```

```python
ar.ID.info()
cr.ID.info()
print (ar.shape)
print (cr.shape)
```

```python
#df = pd.merge(ar, cr, how="left", on="ID" )
df = pd.merge(ar, cr, how="inner", on=["ID"] )
#df = ar.merge(cr, how='inner', on=['ID'])
df.info()
df.describe()
df
```

```python
# 결측치 Imputing: 범주형 변수의 nan 값은 others로 변경
from sklearn.impute import SimpleImputer
import numpy as np
df.OCCUPATION_TYPE.fillna("others", inplace=True)
#df['OCCUPATION_TYPE'] =df['OCCUPATION_TYPE'].replace(np.nan,'others')   #결측치는 others로 치환
df.info()
```

```python
# status 의 c, x 를 6,7로 replace
print(df.STATUS.value_counts())
df.STATUS.replace({"C":"5", "X":"6"}, inplace=True)
print(df.STATUS.value_counts())
df['STATUS']=df['STATUS'].astype(int)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
for col in df.columns :
    print ( df[col].name, df[col].dtype)

[i for i in df.columns if df[i].dtype != 'O']
```

```python
# 수치형 변수 scaling
from sklearn.preprocessing import StandardScaler
numCol  = [col for col in df.columns if  df[col].dtype != "O"]
numCol
s_df = pd.DataFrame (StandardScaler().fit_transform( df[numCol]), columns=numCol)
```

```python
#split
from sklearn.model_selection import train_test_split
train_test_split(s_df, test_size=0.2)
```
